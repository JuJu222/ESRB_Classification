```python
import pandas as pd
```

```python
dataset = pd.read_csv('datasets/final_esrb.csv')
dataset.head()
```

```python
dataset['rating'].value_counts()
```

```python
# convert descriptors to string dtype
dataset['descriptors'] = dataset['descriptors'].astype(str)
```

```python
# split platform into a list of platform
dataset['platform'] = dataset['platform'].apply(lambda x: x.split(','))

# split descriptors into a list of descriptor
dataset['descriptors'] = dataset['descriptors'].apply(lambda x: x.split(','))
```

```python
df = pd.DataFrame()

df['title'] = dataset['title']
# one hot encoding for platform
df = df.join(dataset['platform'].str.join('|').str.get_dummies())
# one hot encoding for descriptors
df = df.join(dataset['descriptors'].str.join('|').str.get_dummies())
#drop nan column
df = df.drop(columns=['nan'])
df.head()
```

```python
rating = {
    'Adults Only 18+': 'A',
    'Everyone': 'E',
    'Everyone 10+': 'E 10+',
    'Mature 17+': 'M',
    'Teen': 'T'
}

df['rating'] = dataset['rating'].map(rating)

df.columns = df.columns.str.lower().str.replace(' ', '_')
df.columns = df.columns.str.replace('^_', '', regex=True)

df.head()
```
